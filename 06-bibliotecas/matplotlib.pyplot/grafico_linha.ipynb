# Gráfico de Linha

Exemplo de utilização da função `plot()` para criar um gráfico de linha simples.

```python
import matplotlib.pyplot as plt

meses = ["Jan", "Fev", "Mar", "Abr"]
vendas = [100, 150, 120, 200]

plt.plot(meses, vendas)

plt.title("Vendas por mês")
plt.xlabel("Mês")
plt.ylabel("Vendas")

plt.show()
```
