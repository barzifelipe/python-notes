# Gráfico de Barras

Exemplo de utilização da função `plot()` para criar um gráfico de barras simples.

```python
import matplotlib.pyplot as plt

produtos = ["Notebook", "Mouse", "Teclado"]
vendas = [50, 120, 80]

plt.bar(produtos, vendas)

plt.title("Vendas por Produto")
plt.xlabel("Produtos")
plt.ylabel("Quantidade")

plt.show()
```
